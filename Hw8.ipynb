```python
from google.colab import drive
drive.mount('/content/drive')
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
import pandas as pd
df = pd.DataFrame()
```

```python
from pathlib import Path
# Шлях до папки з файлами
folder_path = Path('/content/drive/MyDrive/extracted/data')

if folder_path.is_dir():
    # Проходження по кожній папці
    for folder in folder_path.iterdir():
        # Перевірка, чи це папка
        if folder.is_dir():
            # Проходження по кожному файлу у папці
            for file in folder.iterdir():
                # Перевірка, чи це файл та Перевірка, чи файл не порожній
                if file.is_file() and file.stat().st_size > 0:
                    # Зчитування файлу
                    data = pd.read_csv(file)
                    #Обрахунок ознак
                    time_features = pd.DataFrame()
                    time_features['mean'] = data.mean()
                    time_features['min'] = data.min()
                    time_features['max'] = data.max()
                    time_features['median'] = data.median()
                    time_features['std'] = data.median()
                    df2 = pd.DataFrame(time_features.values.reshape(1,-1),columns=['mean_x','min_x','max_x','mean_y','min_y','max_y','mean_z','min_z','max_z','median_x','median_y','median_z','std_x','std_y','std_z'])
                    df2['class'] = folder.name
                    df = pd.concat([df, df2], ignore_index=True)
```

```python
df
```

```python
# Створення словника, де ключами будуть назви папок, а значеннями - числа від 0 до 3
class_to_class_id = {folder_name: i for i, folder_name in enumerate(df['class'].unique())}
# Додавання нової колонки з числовими значеннями папок
df['class_id'] = df['class'].map(class_to_class_id)
```

```python
df
```

```python
X = df.drop(['class', 'class_id'], axis=1)
y=df["class_id"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
# Навчання моделей
# SVM
svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python
# Випадковий ліс
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

```python
# SVM
svm_predictions = svm_model.predict(X_test)
svm_report = classification_report(y_test, svm_predictions)
```

```python
print("SVM Classification Report:")
print(svm_report)
```

```python
# Випадковий ліс
rf_predictions = rf_model.predict(X_test)
rf_report = classification_report(y_test, rf_predictions)
```

```python
print("Random Forest Classification Report:")
print(rf_report)
```


Результати звіту про класифікацію для моделі Random Forest вказують на високу точність моделі для всіх класів, оскільки всі метрики (precision, recall, f1-score) становлять 1.00 для всіх класів. Це означає, що модель вдало класифікує дані для кожного класу без помилок. Точність, чутливість і f1-оцінка на рівні 1.00 означають, що всі прогнози моделі для кожного класу були правильними.

Таким чином, модель Random Forest має дуже високі показники точності та надійності на цьому наборі даних.


Результати звіту про класифікацію для моделі SVM показують такі показники:

Для класів 0, 1 та 3 ми спостерігаємо високу точність, чутливість і f1-оцінку на рівні 1.00, що свідчить про те, що модель правильно класифікувала дані для цих класів без помилок.
Для класу 2 ми бачимо нижчі значення точності, чутливості і f1-оцінки. Це означає, що модель має деякі труднощі у правильній класифікації деяких прикладів цього класу.
Загальна точність моделі (accuracy) становить 0.99, що також вказує на дуже добрі результати класифікації. Однак макро-середня (macro avg) і зважена середня (weighted avg) оцінки за f1-оцінкою є нижчими, ніж для моделі Random Forest. Це може бути через те, що модель SVM має деякі труднощі з класифікацією деяких класів.

У цілому, модель SVM також має високу точність та надійність, але вона може виявити труднощі у правильній класифікації деяких класів порівняно з моделлю Random Forest.
